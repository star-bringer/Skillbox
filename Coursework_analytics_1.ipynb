```python
import datetime as dt #минимальный, максимальный, средний, медианный возраст студентов

age_list = []

new_students_df = students.dropna()#Опустим все пустые значения
brth_date = [pd.to_datetime(date, format ="%Y-%m-%d", errors ="coerce") for date in new_students_df['birthday']]
#Преобразуем в Timestamp, чтобы впоследствии вычислять уже всё в датах.
for date in brth_date:    #Преобразуем в Datetime и находим возраст.
    date = date.to_pydatetime()
    age = (date.today() - date) // dt.timedelta(days=365.2425)
    age_list.append(age)

age_df = pd.DataFrame({'age': age_list})

print('Минимальный возраст студентов: ', age_df.min())#Максимальный и минимальный возраст выглядят, как аномалии, но, возможно,
print('Максимальный возраст студентов: ', age_df.max())# я где-то допустил логическую ошибку.
print('Средний возраст студентов: ', age_df.mean())# Не помню, что бы мы в каком-нибудь модуле углублялись во время.
print('Медианный возраст студентов: ', age_df.median())# Буду рад, если всё-таки найдётся ошибка.
```

```python
import datetime as dt

module_time_pass = []

for course in main['title'].unique():
    course_for_count = main[main.title == course]
    
    for module in course_for_count['module_number'].unique():
        module_for_count = course_for_count[course_for_count.module_number == module]
        hw_in_the_module = module_for_count[module_for_count.is_homework == True]
        done_hw_in_the_mod = hw_in_the_module[hw_in_the_module.status == 'done']
        
        for student in  done_hw_in_the_mod['student_id'].unique():
            student_done_hw = done_hw_in_the_mod[done_hw_in_the_mod.student_id == student]
            start_hw_date = student_done_hw[student_done_hw.lesson_number == min(student_done_hw['lesson_number'])]
            finish_hw_date = student_done_hw[student_done_hw.lesson_number == max(student_done_hw['lesson_number'])]
            
            time_for_pass = (pd.to_datetime(student_done_hw['finish_date']) 
                           -  pd.to_datetime(student_done_hw['start_date'])) 
            
            mdule_time_pass_dict = {'course': course, 'module': module, 'student_id': student, 
                                    'time_for_pass': time_for_pass.dt.days.values[0]}
            module_time_pass.append(mdule_time_pass_dict)

module_time = pd.DataFrame(module_time_pass)
module_time.head()
```
